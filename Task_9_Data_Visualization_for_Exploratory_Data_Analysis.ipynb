```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd

# Sample data
data = pd.DataFrame({
    'Country': ['USA', 'Canada', 'Germany', 'UK', 'France'],
    'GDP': [21.43, 1.84, 4.42, 2.62, 2.78],
    'Population': [328, 38, 83, 66, 67],
    'Area': [9833517, 9976140, 357022, 243610, 551695],
    'Happiness_Index': [7.2, 7.6, 6.9, 7.4, 6.5]
})

# Scatter plot
fig = px.scatter(data, x='GDP', y='Happiness_Index', text='Country', size='Population', color='Country',
                 title='GDP vs Happiness Index')
fig.update_traces(textposition='top center')

# Bar chart
fig2 = px.bar(data, x='Country', y='Population', color='Country', title='Population by Country')

# Pie chart
fig3 = px.pie(data, values='Area', names='Country', title='Area Distribution by Country')

# Line chart
fig4 = go.Figure()
for country in data['Country']:
    fig4.add_trace(go.Scatter(x=[1, 2, 3, 4], y=data[data['Country'] == country]['GDP'],
                               mode='lines+markers', name=country))
fig4.update_layout(title='GDP Trends by Country', xaxis_title='Year', yaxis_title='GDP')

# Displaying the plots
fig.show()
fig2.show()
fig3.show()
fig4.show()
```
